```python
#myFile = np.genfromtxt('temp.csv', delimiter=',')
myFile = np.genfromtxt('temperature.csv', delimiter=',')
```

```python
h = 105216
c1 = [0]*h;
c2 = [0]*h;
c3 = [0]*h;
c4 = [0]*h;
c5 = [0]*h;
c6 = [0]*h;
c7 = [0]*h;
c8 = [0]*h;
data = [0]*h
```

```python
sdvig = -6;
for i in range(h):
    data[i] = myFile[i+1][6+sdvig]
    c1[i] = myFile[i+1][7+sdvig] #50 m
    c2[i] = myFile[i+1][8+sdvig]
    c3[i] = myFile[i+1][9+sdvig]
    c4[i] = myFile[i+1][10+sdvig]
    c5[i] = myFile[i+1][11+sdvig]
    c6[i] = myFile[i+1][12+sdvig]
    c7[i] = myFile[i+1][13+sdvig]
    c8[i] = myFile[i+1][14+sdvig] #0.3 m
```

```python
hist2 = []
hist6 = []
hist8 = []
for i in range(h):
    if c2[i] > 0:
        hist2.append(c2[i])
    if c6[i] > 0:
        hist6.append(c6[i])
    if c8[i] > 0:
        hist8.append(c8[i])
print(len(hist2))
```

```python
H = 2192 #средние за сутки буду считать
day = [0]*H
cr1 = [0]*H;
cr2 = [0]*H;
cr3 = [0]*H;
cr4 = [0]*H;
cr5 = [0]*H;
cr6 = [0]*H;
cr7 = [0]*H;
cr8 = [0]*H;
```

```python
print(c8[6])
print(type(c8[6]))
s = c8[6]
if (c8[5] > 0):
    print ('yes')
else:
    print ('no')
```

```python
time = 48
propusk = 0
for days in range(H):
    kolvo = [0] * 9
    day[days] = (days + 1)/365 + 2011.5
    for t in range(time):
        #50 m
        if (c1[t + days*time] > 0 ): 
            cr1[days] += c1[t + days*time]
            kolvo[1] += 1
        
        #else:
            #propusk += 1
            #print('1'),
        
        #28 m
        if (c2[t + days*time] > 0 ): 
            cr2[days] += c2[t + days*time]
            kolvo[2] += 1
            
                    
        #19 m
        if (c3[t + days*time] > 0 ): 
            cr3[days] += c3[t + days*time]
            kolvo[3] += 1
        
        #10 m
        if (c4[t + days*time] > 0 ): 
            cr4[days] += c4[t + days*time]
            kolvo[4] += 1
        
        #5 m
        if (c5[t + days*time] > 0 ): 
            cr5[days] += c5[t + days*time]
            kolvo[5] += 1
        
        #2 m
        if (c6[t + days*time] > 0 ): 
            cr6[days] += c6[t + days*time]
            kolvo[6] += 1
           
        # 1 m
        if (c7[t + days*time] > 0 ): 
            cr7[days] += c7[t + days*time]
            kolvo[7] += 1
           
        # 0.3 m
        if (c8[t + days*time] > 0 ): 
            cr8[days] += c8[t + days*time]
            kolvo[8] += 1
                       
    if (kolvo[1] > 0):
        cr1[days] /= kolvo[1]
    else:
        cr1[days] = None
        #print(kolvo[1])   
        
    if (kolvo[2] > 0):
        cr2[days] /= kolvo[2]
    else:
        cr2[days] = None
    
    if (kolvo[3] > 0):
        cr3[days] /= kolvo[3]
    else:
        cr3[days] = None
            
    if (kolvo[4] > 0):
        cr4[days] /= kolvo[4]
    else:
        cr4[days] = None
            
    if (kolvo[5] > 0):
        cr5[days] /= kolvo[5]
    else:
        cr5[days] = None
        
    if (kolvo[6] > 0):
        cr6[days] /= kolvo[6]
    else:
        cr6[days] = None
         
    if (kolvo[7] > 0):
        cr7[days] /= kolvo[7]
    else:
        cr7[days] = None
         
    if (kolvo[8] > 0):
        cr8[days] /= kolvo[8]
    else:
        cr8[days] = None
             
    
#print(propusk)
#print(cr1)
```

# гистограммы

```python
import scipy
import scipy.stats as stats
```

## 50 метров

```python
num_bins = 60
n, bins, patches = plt.hist(cr1, num_bins, facecolor='blue', alpha=0.5, density=1,label=u'50 метров')

mu = np.mean(cr1)
sigma = np.std(cr1)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ), linewidth=3, color='y', label=u'нормальное распределение')
plt.xlabel('Температура, $^{\circ}$С', size=12)
#plt.ylabel('Probability')
#plt.title(r'Histogram of IQ: $\mu=100$, $\sigma=15$')

plt.legend()
plt.savefig('гистограмма ряда усредненных за день температур 50 м.png', dpi=300) 
plt.show()
```

```python
print('медиана:   ',np.median(cr1))
print('дисперсия: ',np.var(cr1))
print('среднее:   ',np.mean(cr1))
print('ст. отклон:',np.std(cr1))

W, p = scipy.stats.shapiro(cr1)
print("\n Shapiro-Wilk test statistic, W:", W, "\n", "p-value:", p)
```
